+ Random generate 20 samples for 5 features

```python
import numpy as np
import pandas as pd
import random

# set random seed
np.random.seed(42)

# generate 20 samples for 5 features
n_samples = 20



# set feature ranges 
feature_ranges = {
    'ADIRF': (0.1, 2.5),    
    'APOA1': (10, 50),      
    'CRISP3': (0.5, 10),    
    'PAX2': (0.01, 0.5), 
    'TFPI2': (1, 20)       
}

# generate realistic random data
data = {}
for feature, (min_val, max_val) in feature_ranges.items():
    data[feature] = np.random.uniform(min_val, max_val, n_samples)

sample_input = pd.DataFrame(data)
sample_input.index = [f"sample_{i}" for i in range(1, 21)]
sample_input
```

+ predict the probability of the sample being responsive to platinum-based chemotherapy

```python
import pickle
with open('GPPS_model.pkl', 'rb') as f:
    GPPS = pickle.load(f)

sample_pred = pd.DataFrame(GPPS.predict_proba(sample_input), index=sample_input.index,columns=['Sensitive_probability','Resistant_probability'])
sample_pred 
```
